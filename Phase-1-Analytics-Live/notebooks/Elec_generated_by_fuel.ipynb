```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
#Step 2: Loading CSV File
file_path = "Elec-generated-by-fuel.csv"
df = pd.read_csv("/content/Elec-generated-by-fuel.csv")
```

```python
#Step 3: Cleaning Column Names
df.columns = [str(col).strip().replace("\n", " ").lower() for col in df.columns]

# Step 4: Droping empty rows
df.dropna(how='all', inplace=True)
```

```python
 #Step 5: Identifying numeric columns (skip first column if it's fuel type)
numeric_cols = df.columns[1:]

# Step 6: Converting numeric columns to float
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
#Step 7: Filling missing values (forward-fill, fallback to 0)
df[numeric_cols] = df[numeric_cols].fillna(method='ffill').fillna(0)
```

```python
#Step 8: Removing outliers using SciPy Z-score
z_scores = np.abs(stats.zscore(df[numeric_cols], nan_policy='omit'))
df = df[(z_scores < 3).all(axis=1)]

#Step 9: Reseting index
df.reset_index(drop=True, inplace=True)
```

```python
#Step 10: Exporting to Excel
df.to_excel("Cleaned_Elec_Generated_by_Fuel.xlsx", index=False)

# Step 11: Downloading the file
from google.colab import files
files.download("Cleaned_Elec_Generated_by_Fuel.xlsx")
```
